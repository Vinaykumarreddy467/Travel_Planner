# 🧳 AI Travel Planner

Interactive travel planner powered by:
- **Groq** — Fast LLM for itinerary generation
- **OpenWeatherMap** — Real weather forecasts
- **Google Places API** — Real attractions & restaurants

---

```python
# === SETUP ===
# Make sure you have your .env file set up with your API keys!
from src.planner import TravelPlanner
from src.weather import get_forecast, weather_emoji
from src.places import get_top_attractions, get_top_restaurants

planner = TravelPlanner()

# Check which APIs are configured
missing = planner.check_ready()
if missing:
    print("⚠️  Missing API keys:")
    for key in missing:
        print(f"   • {key}")
    print("\n📝 Edit your .env file to add them, then restart this notebook.")
else:
    print("✅ All APIs configured! Ready to plan trips. 🎉")
```

## 🗺️ Plan Your Trip

Edit the variables below and run the cell to generate a complete itinerary!

```python
# === YOUR TRIP DETAILS ===
# Edit these and run the cell ⬇️

DESTINATION = "Tokyo"           # Where do you want to go?
DAYS = 4                        # How many days?
BUDGET = 2000                   # Total budget in USD (0 = no limit)
PREFERENCES = "foodie, culture, anime, off the beaten path, budget-friendly"

# Generate the plan!
plan = planner.plan_trip(
    destination=DESTINATION,
    days=DAYS,
    budget=BUDGET,
    preferences=PREFERENCES
)

# Print beautiful itinerary
planner.print_itinerary(plan)

# Save to JSON file
saved = planner.save_itinerary(plan, f"output/{DESTINATION.lower().replace(' ', '_')}_itinerary.json")
print(f"📁 Itinerary saved to: {saved}")
```

## 🌤️ Check Weather for Any City

See what the weather looks like before you plan.

```python
# === WEATHER CHECK ===
city = "Paris"  # Change this!
forecast = get_forecast(city, days=5)

if forecast:
    print(f"🌤️  Weather forecast for {city}:\n")
    for w in forecast:
        emoji = weather_emoji(w.condition)
        print(f"  {emoji}  {w.date}:  {w.temp_high}°C / {w.temp_low}°C  -  {w.description}")
else:
    print("⚠️  Could not fetch weather. Is OWM_API_KEY set?")
```

## 📍 Discover Places

Search for real attractions, restaurants, and hotels in any city.

```python
# === PLACES SEARCH ===
city = "Bangkok"  # Change this!

print(f"📍 Top Attractions in {city}:\n")
attractions = get_top_attractions(city)
for i, place in enumerate(attractions, 1):
    stars = "⭐" * int(place.rating or 0)
    print(f"  {i}. {place.name}")
    print(f"     {stars} {place.rating or 'N/A'}")
    print(f"     📍 {place.address}")
    print()

print(f"🍽️  Top Restaurants in {city}:\n")
restaurants = get_top_restaurants(city)
for i, place in enumerate(restaurants, 1):
    stars = "⭐" * int(place.rating or 0)
    dollars = "💰" * (place.price_level or 0)
    print(f"  {i}. {place.name} {dollars}")
    print(f"     {stars} {place.rating or 'N/A'}")
    print(f"     📍 {place.address}")
    print()
```

## 🎯 Quick Trip (One-shot)

Run this cell to quickly plan a trip in one go.

```python
def quick_plan(city, days, budget=0, preferences=""):
    """Plan a trip in one function call."""
    plan = planner.plan_trip(city, days, budget, preferences)
    planner.print_itinerary(plan)
    path = planner.save_itinerary(plan)
    print(f"📁 Saved: {path}")
    return plan

# Example: Quick 3-day trip to Rome
# quick_plan("Rome", 3, 1500, "history, food, photography")

# 🖊️ Edit and uncomment the line above to try it!
```

## 🔧 Troubleshooting

| Problem | Fix |
|---|---|
| `GROQ_API_KEY` missing | Get a free key at https://console.groq.com |
| `OWM_API_KEY` missing | Sign up at https://home.openweathermap.org |
| `GOOGLE_MAPS_API_KEY` missing | Enable Places API in Google Cloud Console |
| Groq returns errors | Try a different model in `src/config.py` |
| Weather returns empty | API key takes ~2 hours to activate after signup |
